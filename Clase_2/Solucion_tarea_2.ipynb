```python
import csv

file_path = "/content/mnist_version_corta.csv"

with open(file_path, newline="") as f:
    reader = csv.reader(f)
    # Convertimos el iterador en una lista para poder manipularlo
    data = list(reader)

num_rows = len(data)
num_cols = len(data[0]) if num_rows > 0 else 0

print(f"Dataset shape: ({num_rows}, {num_cols})")

print("First 5 rows:")
for row in data[:5]:  # Muestra las primeras 5 filas
    print(row)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# el primer numero es 7, el segundo es 2...
first_row = data[0]

# Extraer la etiqueta y los píxeles
first_label = int(first_row[0])  # La primera columna es la etiqueta (el dígito)

# Convertir los píxeles a enteros y guardarlos en una lista
first_image = [int(pixel) for pixel in first_row[1:]]

# Convertir la lista a un array de NumPy
first_image = np.array(first_image, dtype=np.uint8)

# Función para mostrar la imagen
def show_digit(image_vector, label):
    image_matrix = image_vector.reshape(28, 28)  # Convertir a matriz 28x28
    plt.imshow(image_matrix, cmap="gray_r")
    plt.title(f"Digit: {label}")
    plt.axis("off")
    plt.show()

# Mostrar la primera imagen
show_digit(first_image, first_label)
```

```python
# Crear listas para etiquetas y píxeles
labels = []
pixels = []

for row in data:
    labels.append(int(row[0]))  # Guardamos la etiqueta (número de la imagen)
    pixels.append([int(p) for p in row[1:]])  # Guardamos los píxeles como números

# Convertimos los píxeles en un array de NumPy
pixels_array = np.array(pixels)

# Calcular estadísticas de los píxeles
pixel_mean = np.mean(pixels_array)  # Media de los píxeles
pixel_std = np.std(pixels_array)  # Desviación estándar

print(f"Promedio de intensidad de píxeles: {pixel_mean:.2f}")
print(f"Desviación estándar de píxeles: {pixel_std:.2f}")

# Contar la cantidad de veces que aparece cada dígito
label_counts = {}
for label in labels:
    if label not in label_counts:
        label_counts[label] = 0
    label_counts[label] += 1

# Mostrar la cantidad de ejemplos por cada número
print("\nCantidad de imágenes por dígito:")
for digit in sorted(label_counts):
    print(f"Dígito {digit}: {label_counts[digit]}")
```

```python
# parte 5
def caracter(pixel):
    rangos = {
        (0, 64): " ",
        (64, 128): ".",
        (128, 192): "*",
        (192, 256): "#"
    }
    for (a, b), ch in rangos.items():
        if a <= pixel < b:
           return ch
print(caracter(0))
print(caracter(65))
print(caracter(129))
print(caracter(250))
```

```python
digit = 129

# Convertamos los píxeles en caracteres
# conjunto[digit] es una lista de 784 valores (0 a 255) que representan la imagen.
# map(caracter, conjunto[]) convierte cada píxel en un carácter usando la función caracter(pixel).
chars = list(map(caracter, conjunto[digit]))

print(chars[:20])
```

```python
#cuento cuántos píxeles son oscuros en todas las imágenes de un dígito específico.
#recorro todas las imágenes y uso zip() para emparejar con sus etiquetas.
#ignoro imágenes que no sean del dígito deseado.
#uso enumerate() para recorrer cada píxel de la imagen y evalúa si es oscuro (>=128).
#devuelvo una lista de 784 valores indicando la frecuencia de píxeles oscuros en cada posición.

# cuenta píxeles oscuros en imágenes de un dígito específico
def contar_pixeles_negros(conjunto, etiquetas, digito):
    X = [0] * 784  # ya lo vimos
    for valores, etiqueta in zip(conjunto, etiquetas):
        if etiqueta != digito:
            continue
        for i, pixel in enumerate(valores):
            if pixel >= 128:  # 128 es valor umbral
                X[i] += 1  # contador
    # resultado es una lista con la cantidad de veces que cada píxel fue oscuro en las imágenes del dígito dado
    return X

print(contar_pixeles_negros(conjunto, etiquetas, 0))
print(contar_pixeles_negros(conjunto, etiquetas, 1))

# contamos los píxeles oscuros en imágenes del dígito 0
CERO = contar_pixeles_negros(conjunto, etiquetas, 0)
UNO  = contar_pixeles_negros(conjunto, etiquetas, 1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.title("Cero")
sns.heatmap(np.reshape(CERO, (28, 28)), cmap='binary')
plt.figure()
```

```python
plt.title("Uno")
sns.heatmap(np.reshape(UNO, (28, 28)), cmap='binary')
plt.figure()
```

```python
DIF = []
for i in range(len(CERO)):
    DIF.append(abs(CERO[i] - UNO[i]))

#DIF = [abs(CERO[i] - UNO[i]) for i in range(len(CERO))]


plt.title("Diferencia")
sns.heatmap(np.reshape(DIF, (28, 28)), cmap='binary')
plt.figure()
```
